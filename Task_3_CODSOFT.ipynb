Preprocess the data

```python
df.head()
```

```python
df.drop(labels=["RowNumber","CustomerId","Surname"],axis=1,inplace=True)
```

```python
df = pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
x=df.drop(labels=["Exited"],axis=1)
y=df["Exited"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sclar = StandardScaler()
x_train = sclar.fit_transform(x_train)
x_test = sclar.transform(x_test)
```

Train Logistic Regresion model

```python
Ir_model = LogisticRegression()
Ir_model.fit(x_train,y_train)
```

```python
Ir_predictions = Ir_model.predict(x_test)
```

```python
print("Logistic Regresion Model Performance")
print("Confusion Matrix")
print(confusion_matrix(y_test,Ir_predictions))
print("Classification Report")
print(classification_report(y_test,Ir_predictions))
print("Accuracy Score:")
print(accuracy_score(y_test,Ir_predictions))
print("r2_Score:\n",r2_score(y_test,Ir_predictions))
print("Precision Score:",precision_score(y_test,Ir_predictions))
print("Recall Score",recall_score(y_test,Ir_predictions))
print("f1 Score",f1_score(y_test,Ir_predictions))
```

Train Random Forest Model

```python
rf_model = RandomForestClassifier(n_estimators=1000,n_jobs = -1, random_state=42)
rf_model.fit(x_train,y_train)
```

```python
rf_predictions = rf_model.predict(x_test)
```

```python
print("Random Forest Model Performance")
print("Confusion Matrix")
print(confusion_matrix(y_test,rf_predictions))
print("Classification Report")
print(classification_report(y_test,rf_predictions))
print("Accuracy Score:")
print(accuracy_score(y_test,rf_predictions))
print("r2_Score:\n",r2_score(y_test,rf_predictions))
print("Precision Score:",precision_score(y_test,rf_predictions))
print("Recall Score",recall_score(y_test,rf_predictions))
print("f1 Score",f1_score(y_test,rf_predictions))
```

Train Gradient Boosting Model

```python
gb_model = GradientBoostingClassifier(n_estimators=1000,learning_rate = 0.2,random_state=42)
gb_model.fit(x_train,y_train)
```

```python
gb_predictions = gb_model.predict(x_test)
```

```python
print("Gradient Boosting Model Performance")
print("Confusion Matrix")
print(confusion_matrix(y_test,gb_predictions))
print("Classification Report")
print(classification_report(y_test,gb_predictions))
print("Accuracy Score:")
print(accuracy_score(y_test,gb_predictions))
print("r2_Score:\n",r2_score(y_test,gb_predictions))
print("Precision Score:",precision_score(y_test,gb_predictions))
print("Recall Score",recall_score(y_test,gb_predictions))
print("f1 Score",f1_score(y_test,gb_predictions))
```
